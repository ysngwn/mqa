# Por Ano

```python
df = pd.read_csv('doutores_processado_1000.csv', encoding='UTF-8')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.drop('ListaFiliacoes', axis=1, inplace=True)
df.drop('UltimaFiliacao', axis=1, inplace=True)

df = df.astype({
    'IDLattes': int,
    'DuracaoDoutorado': float,
    'Genero': str,
    'GrandesAreas': str,
    'ExclusividadeDoutorado': str,
    'Publicacoes': float,
    'IdadeAcademica': str,
    'HistoricoFiliacoesEmIESInternacional': int,
    'UltimaFiliacaoEmIES': int,
})
```

```python
df_doutores = df.copy(deep=True)

df_doutores_1_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 1.0]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 1.0].index, inplace=True)

df_doutores_2_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 2.0]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 2.0].index, inplace=True)

df_doutores_25_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 2.5]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 2.5].index, inplace=True)

df_doutores_3_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 3.0]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 3.0].index, inplace=True)

df_doutores_35_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 3.5]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 3.5].index, inplace=True)

df_doutores_4_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 4.0]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 4.0].index, inplace=True)

df_doutores_45_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 4.5]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 4.5].index, inplace=True)

df_doutores_5_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 5.0]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 5.0].index, inplace=True)

df_doutores_6_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 6.0]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 6.0].index, inplace=True)

df_doutores_7_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 7.0]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 7.0].index, inplace=True)

df_doutores_8_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 8.0]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 8.0].index, inplace=True)

df_doutores_9_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 9.0]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 9.0].index, inplace=True)

df_doutores_10_ano = df_doutores[df_doutores['DuracaoDoutorado'] == 10.0]
df_doutores.drop(df_doutores[df_doutores['DuracaoDoutorado'] == 10.0].index, inplace=True)

doutores_por_ano = [
    df_doutores_1_ano,
    df_doutores_2_ano,
    df_doutores_25_ano,
    df_doutores_3_ano,
    df_doutores_35_ano,
    df_doutores_4_ano,
    df_doutores_45_ano,
    df_doutores_5_ano,
    df_doutores_6_ano,
    df_doutores_7_ano,
    df_doutores_8_ano,
    df_doutores_9_ano,
    df_doutores_10_ano,
]
```
