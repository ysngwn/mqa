# Por Ano

```python
df = pd.read_csv('doutores_processado_2000_com_todos_int.csv', encoding='UTF-8')
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df['PrimeiraArea'] = df['GrandesAreas'].str.split(';').str[0]
df['PrimeiraArea']
```

```python
# Create a mapping dictionary
unique_areas = df['PrimeiraArea'].unique()
area_to_numeric = {area: idx for idx, area in enumerate(unique_areas)}

# Map the areas to numeric values
df['PrimeiraAreaNumeric'] = df['PrimeiraArea'].map(area_to_numeric)

print(area_to_numeric)
```

```python
df.drop('ListaFiliacoes', axis=1, inplace=True)
df.drop('UltimaFiliacao', axis=1, inplace=True)
df.drop('GrandesAreas', axis=1, inplace=True)
df.drop('PrimeiraArea', axis=1, inplace=True)
```

```python
df.to_csv("doutores_2000_com_area_numerico")
```
